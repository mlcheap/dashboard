```python
import pandas as pd
import sqlite3
con = sqlite3.connect('data/cache.db')

# occupations table
occ = pd.read_sql("SELECT preferredLabel,altLabels,description FROM occupations",con)

# tag table
tags = pd.read_sql("SELECT occupation_title,occupation_id,labeler_id,task_id FROM tags",con)

# task dat table 
tasks = pd.read_sql("SELECT project_name,_id,title,description FROM task_data",con)
```

```python
tags
```

```python
occ = pd.read_sql("SELECT preferredLabel FROM occupations WHERE lang='en'",con)
tasks = tasks.loc[tasks.project_name=='GBR']
df = pd.merge(tasks,tags,left_on='_id',right_on='task_id')
df
```

```python
df2 = pd.merge(df,occ,left_on='occupation_title',right_on='preferredLabel',how='inner')
len(df2)/len(df)
```
